```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv("C:\\datasets\\recommendation engeen\\movies.csv")
movies
```

```python
movies.isnull().sum()
```

```python
#repalacing the null values with null string

movies["genres"]=movies["genres"].fillna('')
```

**TF-IDF Vector**
Term Frequency 
Inverse data Frequency

```python
#converting the text data to feature vectors

from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
feature_vectors=vectorizer.fit_transform(movies["genres"])
print(feature_vectors)
```

**Cosine similarity** means correlation between the vectors



getting the similarity scoes using cosine similarity


```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(feature_vectors)
print(similarity)
```

```python
 print(similarity.shape)
```

**creating a list with all the movie names given in the dataset**

```python
list_of_all_titles= movies['title'].tolist()
print(list_of_all_titles)
```

#  **Getting the movie name from the User**

```python
#getting the movie name from the user

movie_name=input('Enter your favourite movie name: ')
```

```python
#finding the close match for the movie name given by the user

import difflib
find_close_match=difflib.get_close_matches(movie_name,list_of_all_titles)
print(find_close_match)
```

```python
close_match=find_close_match[0]
print(close_match)
```

```python
#finding the index of the movie with title

index_of_the_movie=movies[movies.title==close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
#getting a list of simillar movies

similarity_score= list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
#sorting the movies base on their similarity score

sorted_similar_movies=sorted(similarity_score,key= lambda x:x[1], reverse =True)
print(sorted_similar_movies)
```

**Print the name of similar movies based on the index**

```python
print('Movies suggested for you : \n')
i= 1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index =movies[movies.index==index]['title'].values[0]
    if(i<=10):
        print(i,'.',title_from_index)
    i+=1
```
